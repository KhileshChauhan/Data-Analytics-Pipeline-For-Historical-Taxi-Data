## get file names in bucket dic-taxi-fare-prediction

```pyspark
def getFilenames():
    bucket = 'dic-taxi-fare-prediction'
    prefix = "trip data"

    s3 = boto3.client("s3")
    objects = s3.list_objects(Bucket=bucket, Prefix=prefix)
    tripdata_files = []

    if objects["ResponseMetadata"] and objects["ResponseMetadata"]["HTTPStatusCode"] == 200:
        for obj in objects['Contents']:
            if "Key" in obj:
                tripdata_files.append(obj["Key"])
        print("Found {} under path s3://{}/{}".format(len(tripdata_files), bucket, prefix))
    else:
        print("Could not read files under s3://{}/{}. Exiting.".format(bucket, prefix))

    return tripdata_files

cleanedFiles = getFilenames()
```

## write files to s3

```pyspark
def writeDFtoS3(model):
    # Upload CSV to S3
    bucket = 'dic-taxi-fare-prediction'
    key = 'models'
    s3 = boto3.client("s3")
    s3.put_object(Bucket=bucket, Key=key, Body=model)
```

## Create dataframe

```pyspark
dfs = []
for filename in cleanedFiles:
    bucket = 'dic-taxi-fare-prediction'
    s3Path = "s3n://{}/{}".format(bucket, filename)
    df = pd.read_csv(s3Path,index_col=None, header=0)
    dfs.append(df)

frames = pd.concat(dfs, axis = 0, ignore_index = True)
```
